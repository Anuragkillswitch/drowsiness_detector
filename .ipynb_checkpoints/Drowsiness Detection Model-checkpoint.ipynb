```python
import os    
os.environ['KMP_DUPLICATE_LIB_OK']='True'
import torch
from matplotlib import pyplot as plt
import numpy as np
import cv2  
```

```python
## Now we will load our model 
model=torch.hub.load('ultralytics/yolov5','yolov5s') ##This would load the pretrained Ultralytics model from TorchHub
```

```python
model
```

```python
img = 'https://media.cntraveler.com/photos/53e2f41cdddaa35c30f66775/master/pass/highway-traffic.jpg'
```

```python
results=model(img)
results.print()
```

```python
##Now we need to render this visually using matplotlib
%matplotlib inline 
plt.imshow(np.squeeze(results.render()))  
plt.show()
```

```python
cap=cv2.VideoCapture(0) ##Accessing our webcam
print(cap)
while cap.isOpened(): #Loop through the video
    ret,frame=cap.read()
    cv2.imshow('Drowsiness Detection', frame)
    
    if cv2.waitkey(10) & 0xFF==ord('q'): ##Exiting out this window gracefully ( q is for quit)
        break
        
cap.release() ##release capture device
cv2.destroyAllWindows() 
```
